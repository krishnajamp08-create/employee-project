```python
import pandas as pd
import numpy as np

np.random.seed(0)

n = 458

names = [f"Emp_{i}" for i in range(1, n+1)]
teams = np.random.choice(['HR', 'Sales', 'IT', 'Finance', 'Marketing'], n)
positions = np.random.choice(['Manager', 'Executive', 'Analyst', 'Clerk'], n)
ages = np.random.randint(22, 60, n)
salaries = np.random.randint(30000, 120000, n)
heights = np.random.randint(140, 200, n)  # will fix later

df = pd.DataFrame({
    'Name': names,
    'Team': teams,
    'Position': positions,
    'Age': ages,
    'Salary': salaries,
    'Height': heights
})

df.to_csv('employee_data.csv', index=False)

df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
import numpy as np

# Replace missing or incorrect heights
df['height'] = np.random.randint(150, 181, size=len(df))
```

```python
team_count = df['Team'].value_counts()
team_percent = df['Team'].value_counts(normalize=True) * 100

print(team_count)
print(team_percent)
```

```python
position_count = df['Position'].value_counts()
print(position_count)
```

```python
df['Age_group'] = pd.cut(df['Age'],
                        bins=[20,30,40,50,60],
                        labels=['20-30','30-40','40-50','50-60'])

print(df['Age_group'].value_counts())
```

```python
highest_salary = df.loc[df['Salary'].idxmax()]
print(highest_salary[['Team','Position','Salary']])
```

```python
import matplotlib.pyplot as plt
```

```python
team_count.plot(kind='bar')
plt.title("Employees per Team")
plt.xlabel("Team")
plt.ylabel("Count")
plt.show()
```

```python
position_count.plot(kind='bar')
plt.title("Employees by Position")
plt.show()
```

```python
df['Age_group'].value_counts().plot(kind='bar')
plt.title("Age Group Distribution")
plt.show()
```

```python
df['Salary'].plot(kind='hist')
plt.title("Salary Distribution")
plt.show()
```

```python
plt.scatter(df['Age'], df['Salary'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```
